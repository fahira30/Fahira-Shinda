Pie Chart

```python
#PIE CHART
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal') # Memastikan pie chart berbentuk lingkaran

# Data produk dan jumlah terjual
stock = ['Cheese', 'Bread', 'Milk', 'Ice Cream', 'Noodles', 'Egg']
sold = [10, 50, 45, 15, 37, 113]

# Membuat pie chart
ax.pie(sold, labels=stock, autopct='%1.2f%%')

# Membuat aspek rasio agar tidak ada peringatan
plt.gca().set_aspect('equel', adjustable='box')

# Menyesuaikan aspek agar tidak terjadi peringatan
ax.set_aspect('auto')

plt.show()
```

```python
# Data untuk tiga kategori
akomodasi = [2000, 1100, 4180, 5800, 6420, 980]
kamar = [1503, 10732, 3555, 5206, 2187, 4731, 522]
kasur = [9257, 16415, 6045, 7832, 2885, 8080, 786]
 
# Membuat figure dan axis
fig, ax = plt.subplots(figsize=(8,6))  # Ukuran lebih jelas
 
# Membuat boxplot dengan berbagai pengaturan
box = ax.boxplot([akomodasi, kamar, kasur],
                 vert=False,  # Membuat boxplot horizontal
                 showmeans=True,  # Menampilkan nilai mean
                 meanline=True,  # Menampilkan mean sebagai garis
                 tick_labels=['Akomodasi', 'Kamar', 'Kasur'],  # Perbaikan: 'labels' → 'tick_labels'
                 patch_artist=True,  # Memberikan warna pada boxplot
                 medianprops={'linewidth': 2, 'color': 'blue'},  # Warna garis median
                 meanprops={'linewidth': 2, 'color': 'red'})  # Warna garis mean
 
# Menambahkan warna untuk setiap box
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
 
# Menyesuaikan aspek agar tidak terjadi peringatan
ax.set_aspect('auto')
 
# Menampilkan plot
plt.show()
```

## Scatter PLOT

```python
## Scatter Plot
 
# Data nilai tugas, UTS, dan UAS
tugas = [58, 80, 65, 79, 82]
uts = [78, 80, 85, 59, 92]
uas = [62, 79, 87, 77, 84]
 
# Membuat figure dan axis
fig, ax = plt.subplots()
 
# Scatter plot hubungan tugas dengan UTS
ax.scatter(tugas, uts, color='red', label='Tugas vs UTS')
# Scatter plot hubungan tugas dengan UAS
ax.scatter(tugas, uas, color='blue', label='Tugas vs UAS')
 
# Menambahkan label sumbu dan judul
tempX = 'Tugas'
tempY = 'Nilai'
ax.set_xlabel(tempX)
ax.set_ylabel(tempY)
ax.set_title('Scatter Plot')
 
# Menampilkan legenda
ax.legend()
 
# Menampilkan plot
plt.show()
```

```python
# Mengisi nilai missing value dengan ketentuan:
# Modus untuk data kategorik dan mean untuk data numerik
 
from statistics import mode
 
# Menghitung modus untuk data kategorik
jenis_kelamin_mode = mode(df['jenis_kelamin'].dropna())
pekerjaan_mode = mode(df['pekerjaan'].dropna())
is_keturunan_mode = mode(df['is_keturunan'].dropna())
sampo_mode = mode(df['sampo'].dropna())
 
# Menghitung mean untuk data numerik
berat_mean = df['berat'].mean()
gaji_mean = df['gaji'].mean()
 
# Daftar kolom yang memiliki nilai missing
kolom_missing = ['jenis_kelamin', 'pekerjaan',
                 'is_keturunan', 'berat', 'sampo', 'gaji']
 
# Daftar nilai yang akan digunakan untuk mengisi missing values
inputan = [jenis_kelamin_mode, pekerjaan_mode,
           is_keturunan_mode, berat_mean, sampo_mode, gaji_mean]
 
# Mengisi nilai missing dengan metode yang benar
for kolom, nilai in zip(kolom_missing, inputan):
    df[kolom] = df[kolom].fillna(nilai)
```
